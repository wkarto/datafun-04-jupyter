## Purpose
The purpose of this notebook is to perform exploratory data analysis (EDA) on the Iris dataset. The EDA will include data inspection, descriptive statistics, data distribution visualization, and initial feature engineering.

## Setup and Imports
In this section, we set up the environment and import necessary libraries.

## Step 1. Data Acquisition
Here, we load the Iris dataset and inspect its structure.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

- matplotlib.pyplot is used for creating static, animated, and interactive visualizations.
- pandas is used for data manipulation and analysis.
- seaborn is used for statistical data visualization.

```python
 # Load the Iris dataset into DataFrame
df = sns.load_dataset('iris')

# Inspect first rows of the DataFrame
print(df.head())
```

- sns.load_dataset('iris') loads the Iris dataset into a DataFrame called df.
- df.head() shows the first few rows of the dataset for initial inspection.

## Step 2. Initial Data Inspection
## This section focuses on examining the basic characteristics of the dataset.

```python
print(df.head(10))
print(df.shape)
print(df.dtypes)
```

- df.head(10) provides a snapshot of the first 10 rows to understand the initial data.
- df.shape shows the dimensions of the DataFrame (number of rows and columns).
- df.dtypes lists the data types of each column, which helps in understanding the nature of data.

## Step 3. Initial Descriptive Statistics
Generate summary statistics for a quick overview of the data distribution.

```python
print(df.describe())
```

- df.describe() provides statistical summaries such as mean, median, standard deviation, and percentiles for numerical columns.

## Step 4. Initial Data Distribution for Numerical Columns
Analyze the distribution of numerical features using histograms.

```python
# Inspect histogram by numerical column
df['sepal_length'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

- df['sepal_length'].hist() creates a histogram for the 'sepal_length' column.
- df.hist() generates histograms for all numerical columns to visualize distributions.

## Step 5. Initial Data Distribution for Categorical Columns
Examine the distribution of categorical variables.

```python
# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

- df['species'].value_counts() shows the count of each category in the 'species' column.
- The loop plots count plots for all categorical columns to visualize their distributions.

## Step 6. Initial Data Transformation and Feature Engineering
Perform transformations to enhance the dataset.

```python
# Renaming a column
df.rename(columns={'sepal_length': 'Sepal Length'}, inplace=True)

# Adding a new column
df['Sepal Area'] = df['Sepal Length'] * df['sepal_width']
```

- df.rename() renames the 'sepal_length' column for better readability.
- df['Sepal Area'] is a new column calculated as the product of 'Sepal Length' and 'sepal_width'.

## Step 7. Initial Visualizations
Create visualizations to uncover patterns and relationships in the data.

```python
sns.pairplot(df, hue='species')
plt.show()
```

- sns.pairplot() creates a grid of scatter plots and histograms to visualize pairwise relationships among features, with different colors for each species.

## Summary and Insights:
Observations: The Iris dataset contains measurements of sepal and petal lengths and widths for three species of Iris flowers. Histograms reveal the distribution of numerical features, such as the variation in sepal lengths and petal widths. Count plots show the distribution of species, providing insights into the class balance.
